# inputs : Number of sides and side length of a regular polygon

# output : area

```python
from math import tan,pi
```

```python
n = int(input('number of sides : '))
a = float(input('side length : '))
```

```python
area = (n * (a ** 2)) / (4 * tan(pi / n))
```

```python
print('area is : ',area)
```
