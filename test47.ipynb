# input : a distance in mile and foot

# output : this distance in meter and kilometer

```python
m_per_mile = 1600
m_per_foot = 0.3
```

```python
foot = float(input('the foot part of distance : '))
mile = float(input('the mile part of distance : '))
```

```python
total = foot * m_per_foot + mile * m_per_mile
```

```python
print(total)
print(total/1000)
```
