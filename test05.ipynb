# input:Wind speed at a height of 10 meters and Temperature

# output : windchill

```python
v = float(input('wind speed : '))
t = float(input('temperature : '))
```

```python
windchill = 13.12 + 0.6215 * t - 11.37 * (v ** 0.16) + 0.3965 * t * (v ** 0.16)
```

```python
print('windchill is : ',int(round(windchill,0)))
```
