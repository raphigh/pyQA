# inputs : birth year & now year 

# outputs : the number of years,monthes,days,hours,minutes,second between them

```python
birth = int(input())
now = int(input())
```

```python
years = now - birth
month = years * 12
days = years * 365
hours = days * 24
minutes = hours * 60
seconds = minutes * 60
```

```python
print(years,month,days,hours,minutes,seconds)
```
