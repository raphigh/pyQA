<div style = "direction:rtl"><h2>دترمینان:</h2></div>

<div style = "direction :rtl"> دترمینان یک ماتریس مانند$A\in R^{n \times n} $ یک تابع از مجموعه ماتریس های مربعی است به اعداد حقیقی که با det(A) نمایش داده می شود و از رابطه بازگشتی زیر محاسبه می شود</div>

$\boxed{\textrm{det}(A)=|A|=\sum_{j=1}^n(-1)^{i+j}A_{i,j}|A_{\backslash i,\backslash j}|}
$

<div style = "direction:rtl">در عبارت بالا$\left | A_{\backslash i,\backslash j} \right | $ دترمینان ماتریس حاصل از حذف سطر i ام و ستون j ام می باشد</div>

```python
#in python
def hamsaze(A,i,j):
    for a in range(len(A)):
        B = copy.deepcopy(A)
        B[a].remove(B[a][j])
    B.remove(B[i])
    return B
def det(A):
    sum1 = 0
    for m in range(len(A)):
        sum1 += ((-1)**(m))*(A[0][m])*det(hamsaze(A,0,m))
    return sum1
        
A = [[1, 2, 3],[4, 5, 6],[7, 8, 9]]
det(A)
```

```python
#in numpy
A = np.array([[1,0],[1,2]])
np.linalg.det(A)
```

<div style = "direction:rtl"><h3>تجزیه ی متقارن:یک ماتریس دلخواه A را می توان به صورت حاصل جمع یک ماتریس متقارن و یک ماتریس نامتقارن نوشت</h3></div>

<h3>$َA = \frac{A + A^{T} }{2} + \frac{A - A^{T} }{2} $</h3>

```python
#in python 
def transpose(A):
    B = []
    for i in range(len(A[0])):
        B.append([])
        for j in range(len(A)):
            B[i].append(A[j][i])
    return B

def matsum(A,B):
    C = []
    for i in range(len(A)):
        C.append([])
        for j in range(len(A[1])):
            C[i].append(A[i][j] + B[i][j])
    return C


def matmul(A,c):
    B = []
    for i in range(len(A)):
        B.append([])
        for j in range(len(A[0])):
            B[i].append(c*A[i][j])
    return B


def symdecomp(A):
    print(matmul(matsum(A,transpose(A)),0.5),"+",matmul(matsum(A,matmul(transpose(A),-1)),0.5))
    return "^"
A = [[1,2],[4,5]]
symdecomp(A)
```

```python
#in numpy
A = np.matrix(A)
print((A + np.matrix.transpose(A))/2,"+",(A - np.matrix.transpose(A))/2)
```
