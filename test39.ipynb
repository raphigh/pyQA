# input : an integer in year(named y)

# outputs : y in days,monthes & seconds

```python
y = int(input('insert a number in year please : '))
```

```python
day = y * 365
month = y * 12
second = day * 24 * 60 * 60
```

```python
print('%d years is : '%(y),'\n','%d days'%(day),'\n','%d months'%(month),'\n','%d seconds'%(second))
```
