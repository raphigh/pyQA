# input : radius of a circle

# outputs : diameter,area,perimeter of the circle

```python
from math import pi
```

```python
r = float(input('insert the radius please : '))
```

```python
d = 2 * r
s = pi * (r ** 2)
p = 2 * pi * r
```

```python
print('diameter : ',d)
print('area : ',s)
print('perimeter : ',p)
```
