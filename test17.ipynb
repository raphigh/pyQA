# inputs : now price & interest rate & number of years called n

# output : the price in n years later

```python
price0 = float(input('insert the now price please : '))
interest_rate = float(input('insert the Interest rate please : '))
n = int(input('insert the number of years please : '))
```

```python
future_value = price0 * ((1 + (0.01 * interest_rate)) ** n)
```

```python
print(future_value)
```
