# input : integer n

# outputs : first n sentences of fibonacci sequence

```python
f1 = 1
f2 = 1
```

```python
n = int(input())
```

```python
if n == 1 :
    print('1')
elif n == 2 :
    print('1 1')
else:
    print('1 1',end = ' ')
    for i in range(n - 2):
        a = f2
        f2 = f1 + f2
        f1 = a
        print(f2,end = ' ')
```
