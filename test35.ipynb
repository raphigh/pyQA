# inputs : Gasoline consumption(lit) & distance(mile)

# output : velocity(mile/gal)

note : 1 lit = 0.264179 gal

```python
con = float(input('insert Gasoline consumption in liter please : '))
distance = float(input('insert distance in mile please : '))
```

```python
gal = con * 0.264179
v = distance / gal
```

```python
print(v)
```
