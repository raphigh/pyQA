# inputs : number of pages of a book & the memory of a flash

# output : the number of this book that can be in this flash

note : each page has 30 lines & each line includes 80 bytes

```python
page = int(input('the number of the pages : '))
flash = int(input('the memory of flash : '))
```

```python
lines = page * 30
memory = lines * 80
ans = (flash * ((1024) ** 3))/memory
```

```python
print(ans)
```
