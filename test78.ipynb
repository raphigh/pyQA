# inputs : two natural numbers

# output : compute m ^ n without * operation 

```python
m = int(input())
n = int(input())
```

```python
def prod(x,y):
    p = 0
    for i in range(y):
        p += x
    return p
```

```python
t = m
for i in range(n-1):
    prod(t,m)
    t = prod(t,m)
```
