# input : number of students& id & average of them

# output : rank 2's averang & id of 

```python
n = int(input())
```

```python
av = []
id = []
```

```python
for i in range(n):
    ave = float(input())
    idn = int(input())
    av.append(ave)
    id.append(idn)
for i in range(n - 1):
    for j in range(i + 1,n):
        if av[i] < av[j] :
            a = av[i]
            av[i] = av[j]
            av[j] = a
            a = id[i]
            id[i] = id[j]
            id[j] = a
```

```python
print(av[1])
print(id[1])
```
