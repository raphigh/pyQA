# inputs:radius & height of the cylindrical

# outputs:area & volume of the cylindrical

```python
from math import pi
```

```python
radius = float(input('radius ? '))
height = float(input('height ? '))
```

```python
area = 2 * pi * (radius ** 2) + 2 * pi * radius * height
volume = pi * (radius ** 2) * h
```

```python
print('area is : ',area)
print('volume is : ',volume)
```
