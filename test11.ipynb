# inputs : last year price & this year price

# outputs : Inflation & next year price

```python
last = float(input('insert the last year price please : '))
this = float(input('insert the this year price please : '))
```

```python
inflation = (this - last) / last
next_year = last + this * inflation
```

```python
print('inflation is : ',inflation)
print('next year price is : ',next_year)
```
