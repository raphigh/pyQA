# inputs : two parallel sides of a trapezius(named a & b) & height(named h)

# output : the area of trapezius

```python
s = list(map(float,input('insert two parallel sides of the trapezius : ').split()))
h = float(input('insett the height of the trapezius : '))
```

```python
a = s[0]
b = s[1]
area = ((a + b) / 2) * h
```

```python
print('the area is : ',area)
```
